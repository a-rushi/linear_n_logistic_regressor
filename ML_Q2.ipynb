```python
def comp_err(w,M,N,Y,digit):      
    G = 0
    for i in range(0,N):
        y = np.array(Y[i])
        if y == digit:
            y = 1
        else:
            y = 0
        x = M[i,:]
        x = np.array(x).reshape(785,1)
        w_t = np.array(w.T)
        wx = np.array(w_t@x).reshape(1,1)
        d = 1/(1 + math.exp(wx))
        d_m = 1/(1 + math.exp(-wx))
        g = -y*math.log(d_m) + y*math.log(d) - math.log(d)
        G = G + g
    G = float(1/N)*G
    return G 

def comp_grd(w,M,N,Y,digit):      
    G = 0
    for i in range(0,N):
        y = np.array(Y[i])
        if y == digit:
            y = 1
        else:
            y = 0
        x = M[i,:]
        x = np.array(x).reshape(785,1)
        w_t = np.array(w.T)
        wx = np.array(w_t@x).reshape(1,1)
        n = -y*x + math.exp(wx)*(x - (y)*x)
        d = 1 + math.exp(wx)
        g = (1/d)*n
        G = G + g
    G = float(1/N)*G
    return G 

def log_reg(N,d,M,digit,Y,ita = float(0.1), thrshld = 1000):#,err_thrshld = 0.05,err_imp_thrshld = 0.001):                   
    w_0 = r = np.random.normal(0,0.1,d)
    no_iter = 0
    Cost = []
    w = w_0
    e = float('inf')
    e_bfr = e
    imp = e
    while no_iter < thrshld:# and e > err_thrshld and imp > err_imp_thrshld:
        v = (-1)*comp_grd(w,M,N,Y,digit)
        w = np.array(w).reshape(d,1) + ita*v
        e_bfr = e
        e = comp_err(w,M,N,Y,digit)
        Cost.append(e)
        if e_bfr ==  float('inf'):
            imp = e_bfr
        else:
            imp = e_bfr - e
        no_iter = no_iter+ 1 
    
    Cost = np.array(Cost)
    plt.figure(1)
    plt.plot(np.arange(1, no_iter+1), Cost, label = 'Cost')
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('Cost vs Iterations of Gradient Descent for Digit:' + str(digit))
    plt.show()
    return w 
```

```python
def reg_comp_err(w,M,N,Y,digit,lamda):      
    G = comp_err(w,M,N,Y,digit)
    w_nobias = np.array(w[1:]).reshape(784,1)
    w_sq = w_nobias.T@w_nobias
    g = float(lamda/(2*N))*(w_sq)
    G = G+g
    return G 

def reg_comp_grd(w,M,N,Y,digit,lamda):      
    G = comp_grd(w,M,N,Y,digit)
    w_nobias = np.concatenate(([[0]],np.array(w[1:]).reshape(784,1))) 
    g = float(lamda/N)*(w_nobias)
    G = G + g
    return G 

def reg_log_reg(N,d,M,digit,Y,thrshld,lamda):                   
    w_0 = r = np.random.normal(0,0.1,d)
    no_iter = 0
    Cost = []
    w = w_0
    ita = float(0.4)
    while no_iter < thrshld:
        v = (-1)*reg_comp_grd(w,M,N,Y,digit,lamda)
        w = np.array(w).reshape(d,1) + ita*v
        e = reg_comp_err(w,M,N,Y,digit,lamda)
        Cost.append(e[0][0])
        no_iter = no_iter+ 1 
        
    Cost = np.array(Cost)
    plt.figure(1)
    plt.plot(np.arange(1, no_iter+1), Cost, label = 'Cost')
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('Cost vs Iterations of Regularised Gradient Descent for Digit: ' + str(digit))
    plt.show()
    return w
  
```
